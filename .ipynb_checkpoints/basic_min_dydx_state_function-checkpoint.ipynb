```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.abs(np.random.randn(99))
x = x / np.sum(x) * 100
x_0 = np.array([0])
xx = np.concatenate((x_0,x,x_0))
```

```python
axis= np.arange(np.shape(xx)[0])
plt.scatter(axis,xx)
```

```python
def min_dydx(xx,lr):
    delta = xx[1:] - xx[:100]
    delta_2 = np.abs(delta)
    for i in range(99):
        factor  = delta_2[i] + delta_2[i+1] 
        x_right = xx[i+1] + (factor * lr)
        x_left = xx[i+1] - (factor * lr)
        delta_r = np.abs(x_right - xx[i]) + np.abs(x_right - xx[i+2])
        delta_l = np.abs(x_left - xx[i]) + np.abs(x_left - xx[i+2])
        if delta_r > delta_l:
            if delta_l < delta_2[i] + delta_2[i+1]:
                xx[i+1] = x_left
            else:
                xx[i+1] = xx[i+1]
        else:
            if delta_r < delta_2[i] + delta_2[i+1]:
                xx[i+1] = x_right
            else:
                xx[i+1] = xx[i+1]
    xx = xx / xx.sum() *  100
    return(xx)

def min_dydx_ntime(xx,lr,n):
    for k in range(n):
        delta = xx[1:] - xx[:100]
        delta_2 = np.abs(delta)
        for i in range(99):
            factor  = delta_2[i] + delta_2[i+1] 
            x_right = xx[i+1] + (factor * lr)
            x_left = xx[i+1] - (factor * lr)
            delta_r = np.abs(x_right - xx[i]) + np.abs(x_right - xx[i+2])
            delta_l = np.abs(x_left - xx[i]) + np.abs(x_left - xx[i+2])
            if delta_r > delta_l:
                if delta_l < delta_2[i] + delta_2[i+1]:
                    xx[i+1] = x_left
                else:
                    xx[i+1] = xx[i+1]
            else:
                if delta_r < delta_2[i] + delta_2[i+1]:
                    xx[i+1] = x_right
                else:
                    xx[i+1] = xx[i+1]
        xx = xx / xx.sum() *  100
    return(xx)

```

```python
xx_10 = min_dydx_ntime(xx,0.1,10)
xx_100 = min_dydx_ntime(xx,0.1,100)
xx_1000 = min_dydx_ntime(xx,0.1,1000)

```

```python
plt.plot(np.arange(101),xx)
plt.show()

plt.plot(np.arange(101),xx_10)
plt.show()

plt.plot(np.arange(101),xx_100)
plt.show()

plt.plot(np.arange(101),xx_1000)
plt.show()
```
